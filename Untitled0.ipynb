```python
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('/netflix_titles.csv')
dataset
```

```python

dataset['country'] = dataset['country'].fillna('Unknown')

dataset['country_list'] = dataset['country'].str.split(',')

dataset['country_list'] = dataset['country_list'].apply(lambda x: [c.strip() for c in x])

country = dataset.explode('country_list')

top_countries = country['country_list'].value_counts().head(10)

top_countries.plot(kind='barh', figsize=(10,6))
plt.title("Топ-10 країн у датасеті Netflix")
plt.xlabel("Кількість тайтлів")
plt.ylabel("Країна")
plt.show()
```

```python
dataset['genre_list'] = dataset['listed_in'].str.split(',')
dataset['genre_list'] = dataset['genre_list'].apply(lambda x: [g.strip() for g in x])

genre_exp = dataset.explode('country_list')
genre_exp = genre_exp.explode('genre_list')

for country in ["United States", "India", "Japan"]:
    top_genres = genre_exp[genre_exp['country_list'] == country]['genre_list'].value_counts().head(10)

    top_genres.plot(kind='barh', figsize=(10,6))
    plt.title(f"Топ жанрів у {country}")
    plt.xlabel("Кількість тайтлів")
    plt.ylabel("Жанр")
    plt.show()
```

```python
dataset['is_coproduction'] = dataset['country'].str.contains(',')

copro = dataset.groupby('release_year')['is_coproduction'].mean()

copro.plot(kind='line', figsize=(12,6))
plt.title("Частка копродукцій по роках")
plt.xlabel("Рік")
plt.ylabel("Частка копродукцій")
plt.grid(True)
plt.show()
```
